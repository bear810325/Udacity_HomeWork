```python
# I found that the data type of PatientId is float64, so i change it to int64, the type of AppointmentDay is String, so i convert 
# it to datatime. 
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
df['PatientId'] = df['PatientId'].astype(np.int64)
# I found that some patients made appointment at the same time also at the same hospital. So i think these are duplicated date even though maybe they have 
# different No-show value
df[df.duplicated(subset=['PatientId', 'AppointmentDay', 'Neighbourhood'], keep=False) == True]['AppointmentID'].count()
df.drop_duplicates(subset=['PatientId', 'AppointmentDay', 'Neighbourhood'], keep=False, inplace=True)
# No need to handle null value, because non-null for all columns.
```

```python
# I also drop 'PatiendId', 'AppointmentID', 'ScheduledDay', because i don't think they are associated with No-show feature.
# df.drop(['PatientId', 'AppointmentID', 'ScheduledDay'], axis=1, inplace=True)
# Rename column No-show to NoShow for convenient operation

df.rename(index=str, columns={'No-show': 'NoShow', 'SMS_received':'SMSReceived'}, inplace=True)
df.info()
```

```python
# Create masks for easier graping rows in the future
show = df.NoShow == 'No'
noShow = df.NoShow == 'Yes'
```

```python
# To see if show up is related to SMS_received
df.SMSReceived[show].hist(alpha=0.5, label='show')
df.SMSReceived[noShow].hist(alpha=0.5, label='noShow')
plt.legend();
```

```python
# To see if show up is related to gender
df.groupby('Gender')['NoShow'].value_counts().plot(kind='bar');
```

```python
# To see if show up is related to Scholarship
df.Scholarship[show].hist(alpha=0.5, label='show')
df.Scholarship[noShow].hist(alpha=0.5, label='noShow')
plt.legend();
```
